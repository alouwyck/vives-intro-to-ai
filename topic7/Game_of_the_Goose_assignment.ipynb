Usually, the game is played by more than one player, but here, we keep things simple, and consider one player only. Also, this player uses a single die instead of two dice. The player always starts at square number 1, and the aim of the game is to reach square number 63.

Here, we make this final number variable. This means it should be possible to define a game that contains, for example, 30 squares instead of 63. In this case, the goal is 30 instead of 63!
