### 1.3 Indentation

Indentation is the amount of whitespace (spaces or tabs) before each line of code. **In Python, indentation has meaning!** The indentation determines the grouping of statements.

Consider Java. A group of statements is encapsulated by curly braces.

```java
int x = 5;
if (x < 10) {
System.out.println("x is smaller then 10.");
System.out.println("Good we found that out!");
}
```

Both lines with a print statement form a group of statements, which is delimited by curly braces. It will only be executed if the condition is true (`x < 10`).

In Python, lines with the same indentation are a group of statements (called a suite).

````python
x = 5
if x < 10:
    print("x is smaller then 10.")
    print("Good we found that out!")
if x > 10:
    print("x is bigger then 10.")
print("This line will always be printed.")
````

Both any number of tabs and any number of spaces can be used for indentation, but using 4 spaces per level is recommended.

#### 2.3.3 String indexing


Parts of a string can be addressed using indexing (like an array). **Python uses zero-based indexing!**

Leaving the first index empty will use the start of the string. Leaving the second index empty will use the end of the string.

Negative indexes will start counting from the end, with -1 being the last element.

````python
x = "hello world"
print( x[3]   ) # Index a single character
print( x[0]   )  # Index first character
print( x[-1]  )  # Index last character
print( x[3:8] ) # Index a slice - character with index 8 is NOT included!!
print( x[:8]  ) # Index from the start
print( x[3:]  ) # Index until the end
print( x[-4:] ) # Index the four last characters
print( x[:]   ) # Index from the start until the end
````

#### 2.4.5 List as queues

A queue is a collection to which you only add and take items from the front or the back. This way you can maintain a strict ordening.

For manipulating lists in this way, we may use the following methods:
- `append`
- `extend`
- `pop`
- `insert`

You can add one item using the append and insert method, and multiple items using extend. Pop lets you get an item from a list and at the same time remove it.

````python
lst = [3, 1, 2, 5, 9, 6, 5, 1]
lst.append(10)                  # Add one item to the back of the list
print(lst)

lst.extend([1,2,3,4,5])         # Add multiple items to the back of the list
print(lst)

lst.insert(0, 10)               # Add item to the front of the list (at index 0)
print(lst)

print(lst.pop(0))               # Get and remove an item from the front of the list
print(lst)

print(lst.pop())                # Get and remove an item from the back of the list
print(lst)
````

We can use the list as a fifo queue. FIFO stands for **First In First Out**. The item that was put in first will be taken first. So, the items come out the way you put them in. Just like when a store or restaurant reaches capacity and people line up outside.

````python
fifo = []
fifo.append(5)
fifo.append(3)
print(fifo.pop(0))
fifo.append(9)
fifo.append(10)
print(fifo.pop(0))
print(fifo.pop(0))
print(fifo.pop(0))
````

LIFO, by contrast, stands for **Last In First Out**. Whatever item was put in most recently will be gotten from the queue. Just like a stack of books covering your desk.

````python
lifo = []
lifo.append(5)
lifo.append(3)
print(lifo.pop())
lifo.append(9)
lifo.append(10)
print(lifo.pop())
print(lifo.pop())
print(lifo.pop())
````

### 3.3 For

The for statement is used to iterate over the elements of a sequence (such as a string, tuple or list) or other iterable object. The group of statements (the suite) is then executed once for each item provided by the iterator of the object, in the order returned by the iterator.

In general, an [iterable](https://docs.python.org/3/glossary.html#term-iterable) is an object capable of returning its members one at a time. Examples of iterables include all sequence types (such as list, str, and tuple). In practice, this means you can loop over any sequence object.

One interesting sequence object is the **range**, which will iterate over a sequence of numbers in a range.

````python
print("Range starts at 0 by default")
for x in range(3):  # 3 is not included!!
    print(x)

print()
print("A starting value can be given")
for x in range(5, 10):  # 10 is not included!!
    print(x)

print()
print("The increment can also be defined")
for x in range(0, 10, 2):
    print(x)
````

````python
numbers = [3, 1, 2, 5, 9, 6, 5, 1]
sums = []

for i in range(len(numbers) - 1):
    sums.append(numbers[i] + numbers[i+1])

for x in sums:
    print(x)
````

### 4.2 Defining functions

In Python it is possible to write your own functions. A function may accept one or more input arguments, and it may return one or more output values. It is also possible a function has no input and/or no output.

The general syntax is:

```
def function_name(input1, input2, ...):
    statement(s)
    return output1, output2

```

Note that the body of the function is indented!

````python
# define a function without input and output
def print_hello_world():
    print('Hello, World!')

# call the function
print_hello_world()  # don't forget ()!!
````

````python
# define a function with one input argument
def print_hello(name):
    print('Hello,', name)

# call function
print_hello("Guido")
````

````python
# define a function with two inputs and one output
def sum_of_two(x, y):
    z = x + y
    return z

# call function
print(sum_of_two(y=5, x=3))  # parameter names are required here!
````

````python
# define a function with one input and multiple outputs
def statistics(numbers):
    mean = sum(numbers) / len(numbers)
    return min(numbers), max(numbers), mean

# call function
x = [i for i in range(11)]
min_x, max_x, mean_x = statistics(x)
print("min:", min_x)
print("max:", max_x)
print("mean:", mean_x)
````

##### *Exercise*

Define a function `Fibonacci` that returns the first `n` numbers of the [Fibonnaci sequence](https://en.wikipedia.org/wiki/Fibonacci_number). The function accepts `n` as input argument. Test your function.

###### Solution

````python
def Fibonacci(n):
    fib = [0, 1]
    if n < 2:
        return fib[:n]
    for i in range(2, n):
        fib.append(fib[i-2] + fib[i-1])
    return fib

print(Fibonacci(0))
print(Fibonacci(1))
print(Fibonacci(2))
print(Fibonacci(10))
print(Fibonacci(20))
````

Python has a lot of built-in classes, and it is also possible to define your own classes. We won't go into the specifics of object-oriented programming in Python, and finish this crash course by giving a simple example of how a class and its subclass are defined.

First, consider class `Greeter`:

````python
class Greeter:

    # Constructor
    def __init__(self, name):
        self.name = name  # Create an instance variable (= attribute)

    # Instance method
    def greet(self, loud=False):  # loud is an optional argument
        if loud:
            print('HELLO,', self.name.upper(),'!')
        else:
            print('Hello,', self.name, '!')

    # Instance method
    def length(self):
        return len(self.name)  # return length of attribute name
````

````python
greeter = Greeter('Toby')  # Construct an instance of the Greeter class
greeter.greet()            # Call an instance method; prints "Hello, Toby"
greeter.greet(loud=True)   # Call an instance method; prints "HELLO, TOBY!"
print(greeter.name)        # get 'name' attribute
greeter.name = "Guido"     # change 'name' attribute
greeter.greet()
print(greeter.length())    # print the length of 'name'
````

Note that `__init__` is used as **constructor** and that instance variables or **attributes** are added to the created object `self`. Therefore `self` must be passed as first argument to each instance **method**. As these methods are functions associated with a class instance, they are defined in the same way functions are.

Now consider the following subclass `ProfessionalGreeter` to see how **inheritance** works in Python. Note that `super()` is used to call a superclass method in the subclass:

````python
class ProfessionalGreeter(Greeter):  # ProfessionalGreeter inherits from superclass Greeter

    # Constructor
    def __init__(self, name, title=None):
        super().__init__(name)  # call superclass constructor and pass parameter name
        self.title = title  # Create an attribute for title (attribute name is created by the superclass constructor)

    # Overwrite greet
    def greet(self):
        if self.title is None:
            super().greet()  # call method greet from superclass if no title is given
        else:
            print('Dear', self.title, self.name)
````

````python
professional_greeter = ProfessionalGreeter('Guido', 'Dr.')
professional_greeter.greet()
professional_greeter.title = None
professional_greeter.greet()
````

##### *Exercise*

Define a class `Square`. Its constructor accepts the constant length of the 4 edges, and it assigns this input argument to attribute `length`. Define two instance methods, `perimeter` and `area`, which return the perimeter and the area of the square, respectively. Finally, define a method `print` that prints length, perimeter, and area of an instance of `Square`. Do not recalculate perimeter and area, but use the appropriate methods.

To test your class, instantiate a `Square` object with `length = 5` and print its length, perimeter and area:

```
length = 5
perimeter = 20
area = 25
```



###### Solution

````python
class Square:

    # constructor
    def __init__(self, length):
        self.length = length

    # instance method perimeter
    def perimeter(self):
        return 4 * self.length

    # instance method area
    def area(self):
        return self.length ** 2

    # instance method print
    def print(self):
        print("length =", self.length)
        print("perimeter =", self.perimeter())
        print("area =", self.area())
````

````python
square = Square(length=5)  # instantiate Square object with length = 5
square.print()  # call method print() to display its length, perimeter, and area
````

##### *Exercise*

Define a subclass `Rectangle` that inherits from class `Square`. Its constructor also accepts the constant height of the rectangle, and it assigns this input argument to attribute `height`. Overwrite the two methods, `perimeter` and `area`. Finally, overwrite method `print` that prints length, height, perimeter, and area of an instance of `Rectangle`.

To test your class, instantiate a `Rectangle` object with `length = 5` and `height = 4`, and print its length, height, perimeter, and area:

```
height = 4
length = 5
perimeter = 20
area = 25
```



###### Solution

````python
class Rectangle(Square):  # Rectangle is subclass of Square

    # constructor
    def __init__(self, length, height):
        super().__init__(length)  # first call superclass constructor
        self.height = height

    # overwrite method perimeter
    def perimeter(self):
        return 2 * (self.length + self.height)

    # overwrite method area
    def area(self):
        return self.length * self.height

    # overwrite method print
    def print(self):
        print("height =", self.height)
        super().print()
````

````python
rectangle = Rectangle(length=5, height=4)  # instantiate Rectangle object with length = 5 and height = 4
rectangle.print()  # call method print() to display its height, length, perimeter, and area
````
