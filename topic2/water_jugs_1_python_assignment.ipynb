In the movie "Die Hard with a Vengeance" (Die Hard 3), released in 1995, the water jugs problem is a pivotal part of the plot. John McClane, played by Bruce Willis, and Zeus Carver, played by Samuel L. Jackson, are tasked with solving a complex water jug puzzle as part of a terrorist's twisted game.

The problem is presented as follows. They are given two jugs, one that can hold exactly 3 gallons of water and another that can hold exactly 5 gallons of water. The objective is to measure out exactly 4 gallons of water using only these two jugs and without any other measuring devices.

To successfully solve the puzzle, McClane and Carver must use a combination of filling, pouring, and transferring water between the two jugs while adhering to the constraints of their capacities. Their ability to crack this puzzle is a critical element of the plot, as it leads them to the location of a bomb and plays a significant role in thwarting the terrorist's plans.




The following YouTube video shows the clip in which McClane and Carver are confronted with the problem of the water jugs. The video also explains how the problem can be solved using the "billiard ball" method: https://youtu.be/0Oef3MHYEC0?si=RWZMplWG9RTyMR1p

## 1. Starting by filling the largest jug

First, we start by filling the largest jug of 5 gallons on the left:

As explained in the video, there are three actions required to solve the problem in this way:

1. The horizontal lines going from left to right represent filling the largest jug on the left entirely with water.

2. The diagonal lines going from upper right to lower left represent emptying the smallest jug on the right completely.

3. The diagonal lines going from lower right to upper left represent pouring the water from the largest jug into the smallest jug until the latter is entirely filled with water or the first is emptied completely.

Now we've defined the three operations required to solve the problem, we are able to implement the "billiard ball" algorithm, as explained in the video and shown in the picture. The "billiard table" has four borders, so there are four possibilities which are implemented using an `if` statement:

1. The right border represents the largest jug being full of water, in which case water is poured from this jug into the smallest one.

2. The upper border represents the smallest jug being full of water, in which case it is emptied.

3. The left border represents the largest jug being empty, in which case it is filled entirely with water.

4. The lower border represents the smallest jug being empty, in which case water is poured from the largest jug into the smallest one.
